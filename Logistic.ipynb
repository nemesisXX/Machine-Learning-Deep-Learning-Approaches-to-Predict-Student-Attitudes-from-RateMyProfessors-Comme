Liblinear

```python
score=[]
loss=[]
C=[]
auroc1=[]
for c in range(1,10000,1000):
  C.append(c)
  logit= linear_model.LogisticRegression(C=c,solver='liblinear')
  logit.fit(x_train, y_train)
  predictions = logit.predict(x_test)
  score.append(logit.score(x_test,y_test))
  auroc1.append(sklearn.metrics.roc_auc_score(y_test,predictions))
  loss.append(sklearn.metrics.log_loss(y_test, predictions))
```

```python
import matplotlib.pyplot as plt
plt.ylabel('Score')
plt.xlabel('Regularization')
plt.plot(C,score)
```
