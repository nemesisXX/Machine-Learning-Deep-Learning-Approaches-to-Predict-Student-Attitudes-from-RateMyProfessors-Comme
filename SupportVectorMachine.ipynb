```python
from sklearn.metrics import roc_auc_score
from sklearn import svm
```

```python
C=[8,10,12,14,16]
auroc1=[]
for c in C:
  clf=svm.SVC(C=c,kernel='linear',max_iter=1000)
  clf.fit(x_train,y_train)
  predicted = clf.predict(x_test)
  auroc1.append(roc_auc_score(y_test,predicted))
```

```python
C=[8,10,12,14,16]
auroc2=[]
for c in C:
  clf=svm.SVC(C=c,kernel='poly',max_iter=1000)
  clf.fit(x_train,y_train)
  predicted = clf.predict(x_test)
  auroc2.append(roc_auc_score(y_test,predicted))
```

```python
C=[8,10,12,14,16]
auroc3=[]
for c in C:
  clf=svm.SVC(C=c,kernel='rbf',max_iter=1000)
  clf.fit(x_train,y_train)
  predicted = clf.predict(x_test)
  auroc3.append(roc_auc_score(y_test,predicted))
```

```python
C=[8,10,12,14,16]
auroc4=[]
for c in C:
  clf=svm.SVC(C=c,kernel='sigmoid',max_iter=1000)
  clf.fit(x_train,y_train)
  predicted = clf.predict(x_test)
  auroc4.append(roc_auc_score(y_test,predicted))
```

```python
import matplotlib.pyplot as plt
C=[8,10,12,14,16]
plt.plot(C, auroc1)
plt.plot(C, auroc2)
plt.plot(C, auroc3)
plt.plot(C, auroc4)


plt.title('Support Vector Machine')
plt.xlabel("Regularization Strength")
plt.ylabel("AUROC")
plt.legend(['linear','poly','rbf','sigmoid'], loc='upper right')
plt.rcParams["figure.figsize"] = (10,8)
plt.savefig('svm.png', dpi = 300)
plt.show()
plt.clf()
```
