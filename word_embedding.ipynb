```python
# This is a word embedding process. Here we use sentence tranformer model, which trains based on "BERT". This transformation
# is more precise than our enwiki.model, which used "word2vec".
import faulthandler
import pandas as pd
import numpy as np
from sentence_transformers import SentenceTransformer
import json
faulthandler.enable()
```

```python
def feature_vector(csv_file):
    df = pd.read_csv(csv_file,sep=",")
    df = df.loc[2:,"processed_comments"]
    df.dropna(how = "any",inplace = True)
    lst = list(df)
    model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
    embeddings = model.encode(lst)
    return embeddings
    

# csv_file = sys.argv[1]
csv_file = "processed.csv"
embeddings = feature_vector(csv_file)
```

```python
df = pd.read_csv("processed.csv",sep = ",")
df.drop(index=df.index[:2],inplace=True)
df.dropna(how = "any",inplace = True)
dep = np.expand_dims(np.array(df["department_name"]), axis=1)
sta = np.expand_dims(np.array(df["state_name"]),axis=1)
num = np.expand_dims(np.array(df["num_student"]),axis=1)
lab = np.expand_dims(np.array(df["label"]),axis=1)
stu = np.expand_dims(np.array(df["student_difficult"]),axis=1)
wor = np.expand_dims(np.array(df["word_comment"]),axis=1)
fea_vec = np.array(embeddings)
arr = np.concatenate((dep,sta,num,stu,wor,fea_vec),axis = 1)
arr1 = np.concatenate((dep,sta,num,lab,stu,wor,fea_vec),axis = 1)
```

```python
np.savetxt('feature_matrix.txt',arr1)
content = np.loadtxt('feature_matrix.txt')
print(content.shape)
```

```python
df = pd.DataFrame(data=arr.astype(float))
df.to_csv('outfile.csv', sep=' ', header=False, index=False)
```
